```python
# Week 1 Mini-Project: Python Basics

# Question 1: User Data Collector
def user_data_collector():
    user_data = {}
    user_data['name'] = input("Enter your name: ")
    user_data['age'] = input("Enter your age: ")
    user_data['email'] = input("Enter your email: ")
    user_data['favorite_number'] = input("Enter your favorite number: ")

    if '@' in user_data['email'] and '.' in user_data['email']:
        print(f"\nCollected Data:\nName: {user_data['name']}\nAge: {user_data['age']}\nEmail: {user_data['email']}\nFavorite Number: {user_data['favorite_number']}")
    else:
        print("Invalid email format!")

# Question 2: Even or Odd?
def is_even(number):
    return number % 2 == 0

def check_even_odd():
    num = int(input("Enter a number: "))
    if is_even(num):
        print(f"{num} is even.")
    else:
        print(f"{num} is odd.")

# Question 3: Temperature Converter
def convert_temperature(temp, scale):
    if scale == "C":
        return (temp * 9/5) + 32
    elif scale == "F":
        return (temp - 32) * 5/9
    else:
        return "Invalid scale"

def temperature_converter():
    temp = float(input("Enter temperature: "))
    scale = input("Enter scale (C for Celsius, F for Fahrenheit): ").upper()
    converted = convert_temperature(temp, scale)
    print(f"Converted Temperature: {converted}")

# Question 4: Finding Min & Max
def find_max_min(numbers_list):
    return max(numbers_list), min(numbers_list)

def min_max_finder():
    numbers = [int(input(f"Enter number {i+1}: ")) for i in range(5)]
    maximum, minimum = find_max_min(numbers)
    print(f"Max: {maximum}, Min: {minimum}")

# Question 5: Student Data Manager
def student_data_manager():
    students = []
    for i in range(3):
        name = input(f"Enter student {i+1} name: ")
        age = int(input(f"Enter {name}'s age: "))
        grade = input(f"Enter {name}'s grade: ")
        students.append((name, age, grade))

    student_dict = {name: (age, grade) for name, age, grade in students}
    print("\nStudent Data:")
    for key, value in student_dict.items():
        print(f"{key}: Age - {value[0]}, Grade - {value[1]}")

# Question 6: Inventory Management System
def update_inventory(inventory_dict, item, quantity):
    updated_quantity = max(0, inventory_dict.get(item, 0) + quantity)
    inventory_dict[item] = updated_quantity
    return inventory_dict

def inventory_manager():
    inventory = {
        "apples": 10,
        "bananas": 15,
        "oranges": 20,
        "grapes": 12,
        "mangoes": 8
    }

    for _ in range(3):
        item = input("Enter item to update: ").lower()
        quantity = int(input("Enter quantity to add/remove (use negative to remove): "))
        inventory = update_inventory(inventory, item, quantity)

    print("\nUpdated Inventory:")
    for item, qty in inventory.items():
        print(f"{item}: {qty}")

if __name__ == "__main__":
    user_data_collector()
    check_even_odd()
    temperature_converter()
    min_max_finder()
    student_data_manager()
    inventory_manager()

```
