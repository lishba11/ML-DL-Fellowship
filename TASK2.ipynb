```python
# 1. Mean, Median, and Mode Calculation
def mean(numbers):
    return sum(numbers) / len(numbers)

def median(numbers):
    numbers.sort()
    n = len(numbers)
    mid = n // 2
    return numbers[mid] if n % 2 == 1 else (numbers[mid - 1] + numbers[mid]) / 2

def mode(numbers):
    freq = {}
    for num in numbers:
        freq[num] = freq.get(num, 0) + 1
    max_freq = max(freq.values())
    modes = [num for num, count in freq.items() if count == max_freq]
    return modes if len(modes) > 1 else modes[0]

# Example usage
numbers = [4, 1, 2, 2, 3, 3, 3, 4, 4, 4]
print("Mean:", mean(numbers))
print("Median:", median(numbers))
print("Mode:", mode(numbers))

# 2. Variance Calculation
def variance(numbers):
    m = mean(numbers)
    return sum((x - m) ** 2 for x in numbers) / len(numbers)

print("Variance:", variance(numbers))

# 3. Standard Deviation Calculation
def standard_deviation(numbers):
    return variance(numbers) ** 0.5

print("Standard Deviation:", standard_deviation(numbers))

# 4. Euclidean Distance Calculation
def euclidean_distance(x, y):
    if isinstance(x, list) and isinstance(y, list) and len(x) == len(y):
        return sum((a - b) ** 2 for a, b in zip(x, y)) ** 0.5
    else:
        return abs(x - y) ** 0.5

print("Euclidean Distance (Values):", euclidean_distance(3, 7))
print("Euclidean Distance (Lists):", euclidean_distance([1, 2, 3], [4, 5, 6]))

# 5. Sigmoid Function
def sigmoid(x):
    return 1 / (1 + (2.71828 ** -x))

print("Sigmoid(0):", sigmoid(0))
print("Sigmoid(2):", sigmoid(2))
```
