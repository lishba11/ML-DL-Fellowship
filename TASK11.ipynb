```python
#Filtering Out Missing Data
import pandas as pd
import numpy as np

# Series example
series_data = pd.Series([1, np.nan, 3.5, np.nan, 7])

# Drop missing values using dropna
clean_series = series_data.dropna()

# Same result using boolean indexing with notna
clean_series_alt = series_data[series_data.notna()]

print("Original Series:\n", series_data)
print("\nCleaned Series using dropna:\n", clean_series)
print("\nCleaned Series using notna:\n", clean_series_alt)

# DataFrame example
df = pd.DataFrame([
    [1.0, 6.5, 3.0],
    [1.0, np.nan, np.nan],
    [np.nan, np.nan, np.nan],
    [np.nan, 6.5, 3.0]
])

# Drop rows with any NaNs
dropped_any_na = df.dropna()

# Drop rows where all values are NaN
dropped_all_na = df.dropna(how='all')

# Add a column with all NaNs
df[3] = np.nan

# Drop columns where all values are NaN
dropped_columns_all_na = df.dropna(axis='columns', how='all')

# Keep rows with at least 2 non-NaN values using thresh
df2 = pd.DataFrame(np.random.randn(7, 3))
df2.iloc[:4, 1] = np.nan
df2.iloc[:2, 2] = np.nan
dropped_thresh = df2.dropna(thresh=2)

# Display results
print("\nOriginal DataFrame:\n", df)
print("\nDropped rows with any NaNs:\n", dropped_any_na)
print("\nDropped rows with all NaNs:\n", dropped_all_na)
print("\nDropped columns with all NaNs:\n", dropped_columns_all_na)
print("\nDataFrame with thresh=2:\n", dropped_thresh)
```

```python
#Filling In Missing Data
import pandas as pd
import numpy as np

# Sample DataFrame with missing values
df = pd.DataFrame({
    0: [0.1, -0.2, 0.3, np.nan, 0.5, np.nan],
    1: [np.nan, 0.6, np.nan, np.nan, 0.9, -1.1],
    2: [1.2, np.nan, 0.8, -0.3, np.nan, np.nan]
})

print("Original DataFrame:\n", df)

# Fill missing values with a constant
filled_constant = df.fillna(0)
print("\nFilled with constant 0:\n", filled_constant)

# Fill missing values with different values per column
filled_dict = df.fillna({1: 0.5, 2: 0})
print("\nFilled with column-specific values:\n", filled_dict)

# Forward fill method
filled_ffill = df.fillna(method="ffill")
print("\nForward fill:\n", filled_ffill)

# Forward fill with a limit
filled_ffill_limit = df.fillna(method="ffill", limit=1)
print("\nForward fill with limit=1:\n", filled_ffill_limit)

# Fill with column mean (example using a Series)
s = pd.Series([1., np.nan, 3.5, np.nan, 7])
filled_mean = s.fillna(s.mean())
print("\nSeries filled with mean:\n", filled_mean)
```

```python
#Replacing Values
#Renaming Axis Indexes
import pandas as pd
import numpy as np

# Sample sales dataset
data = pd.DataFrame({
    "product": ["laptop", "tablet", "phone", "laptop", "tablet", "phone", "headphones"],
    "sales": [1000, 500, -999, 1000, -999, 750, 300],
    "region": ["North", "East", "South", "North", "East", "South", "West"]
})

# Display the original data
print("Original Data:")
print(data)

# 1. Replacing -999 with NaN in the 'sales' column
data["sales"] = data["sales"].replace(-999, np.nan)

# Replace NaN values in 'sales' column with the mean of 'sales'
mean_sales = data["sales"].mean()
data["sales"].fillna(mean_sales, inplace=True)

# Display the data after replacing values
print("\nData After Replacing -999 with NaN and filling NaN with mean sales value:")
print(data)

# 2. Renaming the columns and index (row labels)
data.rename(index={"North": "NORTH_REGION", "East": "EAST_REGION", "South": "SOUTH_REGION", "West": "WEST_REGION"},
            columns={"product": "product_name", "sales": "sales_amount", "region": "sales_region"}, inplace=True)

# Display the data after renaming axis indexes
print("\nData After Renaming Index and Columns:")
print(data)

# 3. Function to categorize sales
def categorize_sales(row):
    if row['sales_amount'] > 800:
        return "High Sales"
    elif 500 <= row['sales_amount'] <= 800:
        return "Medium Sales"
    else:
        return "Low Sales"

# Apply the function to categorize sales based on the sales_amount
data["sales_category"] = data.apply(categorize_sales, axis=1)

# Display the final transformed data
print("\nData After Categorizing Sales:")
print(data)
```

```python
# Binning and Discretization:
import pandas as pd
import numpy as np

# Sample sales dataset
sales = [1000, 500, 999, 1200, 800, 750, 300, 1200, 1500, 2000, 1000, 900]
sales_df = pd.DataFrame(sales, columns=["sales"])

# Define bins for sales: you can change the ranges to suit your data
bins = [0, 500, 1000, 1500, 2500]
bin_labels = ["Low", "Medium", "High", "Very High"]

# Use pd.cut to categorize the sales into bins
sales_categories = pd.cut(sales_df["sales"], bins=bins, labels=bin_labels, right=False)

# Add the new category column to the DataFrame
sales_df["sales_category"] = sales_categories

print("Sales Dataset with Categories:")
print(sales_df)
```

```python
#Permutation and Random Sampling and computing indicators and variables
import numpy as np

# Example sales dataset
data = {
    'Product': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Sales': [120, 250, 300, 400, 150, 220],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South']
}

df = pd.DataFrame(data)

# 1. Random Sampling without replacement (3 samples)
sampled_data = df.sample(n=3)
print("Random Sampled Data:")
print(sampled_data)
print("\n")

# 2. Creating Dummy Variables for the 'Region' column
dummies = pd.get_dummies(df['Region'], prefix='Region')
df_with_dummies = df.join(dummies)
print("Data with Dummy Variables:")
print(df_with_dummies)
print("\n")

# 3. Permuting Rows
row_sampler = np.random.permutation(df.shape[0])  # Number of rows
permuted_rows = df.take(row_sampler)
print("Permuted Rows:")
print(permuted_rows)
print("\n")

# 4. Permuting Columns
column_sampler = np.random.permutation(df.shape[1])  # Number of columns
permuted_columns = df.take(column_sampler, axis='columns')
print("Permuted Columns:")
print(permuted_columns)

import numpy as np

# Example sales dataset
data = {
    'Product': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Sales': [120, 250, 300, 400, 150, 220],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South']
}

df = pd.DataFrame(data)

# 1. Random Sampling without replacement (3 samples)
sampled_data = df.sample(n=3)
print("Random Sampled Data:")
print(sampled_data)
print("\n")

# 2. Creating Dummy Variables for the 'Region' column
dummies = pd.get_dummies(df['Region'], prefix='Region')
df_with_dummies = df.join(dummies)
print("Data with Dummy Variables:")
print(df_with_dummies)
print("\n")

# 3. Permuting Rows
row_sampler = np.random.permutation(df.shape[0])  # Number of rows
permuted_rows = df.take(row_sampler)
print("Permuted Rows:")
print(permuted_rows)
print("\n")

# 4. Permuting Columns
column_sampler = np.random.permutation(df.shape[1])  # Number of columns
permuted_columns = df.take(column_sampler, axis='columns')
print("Permuted Columns:")
print(permuted_columns)

import pandas as pd
import numpy as np

# Example sales dataset
data = {
    'Product': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Sales': [120, 250, 300, 400, 150, 220],
    'Region': ['North', 'South', 'East', 'West', 'North', 'South']
}

df = pd.DataFrame(data)

# 1. Random Sampling without replacement (3 samples)
sampled_data = df.sample(n=3)
print("Random Sampled Data:")
print(sampled_data)
print("\n")

# 2. Creating Dummy Variables for the 'Region' column
dummies = pd.get_dummies(df['Region'], prefix='Region')
df_with_dummies = df.join(dummies)
print("Data with Dummy Variables:")
print(df_with_dummies)
print("\n")

# 3. Permuting Rows
row_sampler = np.random.permutation(df.shape[0])  # Number of rows
permuted_rows = df.take(row_sampler)
print("Permuted Rows:")
print(permuted_rows)
print("\n")

# 4. Permuting Columns
column_sampler = np.random.permutation(df.shape[1])  # Number of columns
permuted_columns = df.take(column_sampler, axis='columns')
print("Permuted Columns:")
print(permuted_columns)
```

```python
import pandas as pd
import re

# Sample Sales Data
data = {
    'Product': ['Apple', 'Banana', 'Orange', 'Grapes'],
    'Description': ['Fresh and juicy apple', 'Sweet yellow banana', 'Tasty and fresh orange', 'Green and sweet grapes'],
    'Price': ['10.99', '5.50', '8.75', '12.00']
}

# Create DataFrame
df = pd.DataFrame(data)

# 1. String Manipulation: Convert 'Price' column to float
df['Price'] = df['Price'].astype(float)

# 2. String Manipulation: Replace words in 'Description' column
df['Description'] = df['Description'].str.replace('fresh', 'delicious')

# 3. Regex: Extract product names that contain the letter 'e' from 'Description'
regex_pattern = r"\b\w*e\w*\b"  # Matches words containing the letter 'e'
df['Words_with_e'] = df['Description'].apply(lambda x: re.findall(regex_pattern, x))

# 4. Regex: Find prices greater than 8 and mark them as 'High' or 'Low'
df['Price_Category'] = df['Price'].apply(lambda x: 'High' if x > 8 else 'Low')

# 5. String Manipulation: Concatenate 'Product' and 'Price' into a new column
df['Product_Info'] = df['Product'] + " - " + df['Price'].astype(str)

# Display the DataFrame
print(df)
```
